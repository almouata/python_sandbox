```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,10)
y = np.sin(x)

plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("cool tilte")
plt.show()
```
