```python
import numpy as np
from scipy.stats import multivariate_normal as mvn
import matplotlib.pyplot as plt
```

```python
cov = np.array([[1, 0.8], [0.8, 3]])
mu = np.array([0,2])
r = mvn.rvs(mu, cov=cov, size=1000)
```

```python
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
f = np.random.multivariate_normal(mean=mu,cov=cov,size=1000)
plt.scatter(f[:,0], f[:,1])
plt.axis('equal')
plt.show()
```
